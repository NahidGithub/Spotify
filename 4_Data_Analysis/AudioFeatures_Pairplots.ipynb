```python
import os
import boto3
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
from s3 import get_file
```

# Connect to Amazon S3

```python
s3 = boto3.resource('s3')
get_file(s3, "s3ssp", download_file="NLP_Data/new_master_lyrics_audio_features.csv", rename_file="lyrics_audiofeatures.csv")
```

```python
get_file(s3, "s3ssp", download_file="Analysis_Data/master_audiofeatures_track_uri.csv", rename_file="master_audiofeatures.csv")
```

```python
#read csv into pandas dataframe for personal collection music pair plot
df_master = pd.read_csv("master_audiofeatures.csv", sep='|')
```

```python
#read csv into pandas dataframe for lyrics database pair plot
df_lyrics = pd.read_csv("lyrics_audiofeatures.csv", sep='|')
```

```python
df_master.head()
```

# Pairplots for Audio Features

```python
df_master_sns = df_master[['danceability', 'energy','speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence']]
df_lyrics_sns = df_lyrics[['danceability', 'energy','speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence']]
```

```python
g = sns.pairplot(df_master_sns)
g.fig.suptitle("Personal Music Collection Pairplot", y=1.03, fontsize=30)
```

```python
a = sns.pairplot(df_lyrics_sns)
a.fig.suptitle("Lyrics Database Pairplot", y=1.03, fontsize=30)
```
