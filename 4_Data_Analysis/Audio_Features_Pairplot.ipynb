```python
import pandas as pd
import os
import seaborn as sns
from matplotlib import pyplot
import boto3
from s3 import get_file
```

```python
# Connect to Amazon S3
s3 = boto3.resource('s3')
lyrics = get_file(s3,'s3ssp',download_file='Analysis_Data/master_audiofeatures_track_uri.csv',rename_file='audiofeatures.csv')
lyrics_two = get_file(s3,'s3ssp',download_file='NLP_Data/new_master_lyrics_audio_features.csv',rename_file='audiofeatures.csv')
```

```python
# create Pandas DataFrame
df_master = pd.read_csv(lyrics,sep='|',encoding='utf-8')
df_new = pd.read_csv(lyrics_two,sep='|',encoding='utf-8')
```

```python
#select columns for pairplot
df_master_sns = df_master[['acousticness', 'danceability', 
                           'energy', 'valence', 'instrumentalness', 
                           'liveness', 'speechiness']]
df_new_sns = df_new[['acousticness', 'danceability', 
                     'energy', 'valence', 'instrumentalness', 
                     'liveness', 'speechiness']]
```

# Create Pairplots for selected audio features

```python
sns.pairplot(df_master_sns)
```

```python
sns.pairplot(df_new_sns)
```
